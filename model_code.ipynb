```
labels=file['annotation/label/0\t'].tolist()
count=file['annotation/label/0\t'].count()
count
```

```
print(labels[0:10])
print(len(labels))
```

```
import sklearn.model_selection as model_selection
sentences=file['content'].tolist()
sentences,test_sentences,labels,test_labels=model_selection.train_test_split(sentences,labels, train_size=0.8,test_size=0.2)
```

```
print(sentences[0:3])
print(test_sentences[0:3])
```

```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
```

```
tokenizer=Tokenizer(oov_token="<OOV>")
tokenizer.fit_on_texts(sentences)
```

```
word_index=tokenizer.word_index
print(word_index)
```

```
sequences=tokenizer.texts_to_sequences(sentences)
padded=pad_sequences(sequences,padding='post',maxlen=342)
```

```
test_sequences=tokenizer.texts_to_sequences(test_sentences)
test_padded=pad_sequences(test_sequences,padding='post',maxlen=342)
```

```
print(padded[0])
print(padded.shape)
print(test_padded[0])
print(test_padded.shape)
```

```
padded=np.array(padded)
labels=np.array(labels)
test_padded=np.array(test_padded)
test_labels=np.array(test_labels)
print(labels[0])
```

```
model=tf.keras.Sequential([
                           tf.keras.layers.Embedding(100000,16,input_length=342),
                           
                           #tf.keras.layers.GlobalAveragePooling1D(),
                          tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32)),
                           tf.keras.layers.Dense(6,activation='relu'),
                           tf.keras.layers.Dense(1,activation='sigmoid')
                          ])
```

```
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```
epo=10
```

```
history=model.fit(padded,labels,epochs=epo,validation_data=(test_padded,test_labels),verbose=2)
```
